```python
#TASK1


#What is the total annual salary for all the employees of each department? Display the result using a bar graph.

#Hint:

#1) Summarize the department-wise sum of annual salaries.

#ii) Use index and values properties of the summarized data to plot the bar graphs.
import pandas as pd
import numpy as np
#for visualization
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df=pd.read_excel(r'Downloads/DS1_C4_S5_Employee_Data_Practice.xlsx')
df
```

```python
#DEPARTMENT WISE ANNAUL SALARY
df1 = df.groupby('Department')['Annual Salary ($)'].sum()
df1

plt.bar(df1.index, df1.values, color='pink')
plt.xlabel('Department')
plt.ylabel('Total Annual Salary')
plt.title('Total Annual Salary for Each Department')
plt.xticks(rotation=20)
plt.show()
```

```python
#TASK2
#The company wants to analyze gender equality across several departments. Display the department-wise total number of male and female employees using a DataFrame

#Hints:

#1. Summarize the department- and gender-wise count of employees.

#2. Create a DataFrame with all department names and the respective male and female counts.

df2=pd.DataFrame(df)
df2 = df.groupby(['Department','Gender'])['Gender'].count()
df2
df2.plot(x='Gender',kind='bar',xlabel='Department',ylabel='Gender',stacked=False)


```

```python
#TASK3

df1=pd.DataFrame(df)
filtered_df = df1[(df1['Department'] == 'IT') &  (df['Work_Experience'] == 0)]
filtered_df

# Calculate the average salary of the filtered data
average_salary = filtered_df['Annual Salary ($)'].mean()
average_salary

print(f"The average salary for a new employee with no work experience in the IT department is: ${average_salary:.2f}")
```

```python
#TASK4

df_grouped = df.groupby('Department')['Annual Salary ($)'].sum()

plt.pie(df_grouped, labels=df_grouped.index, autopct='%.2f%%')
plt.title('Department-wise Ratio for Sum of Annual Salaries')
```

```python
#TASK1
import numpy as np
import pandas as pd
df=pd.read_csv(r"Downloads/DS1_C4_S4_Test_Scores_Data_Practice.csv")
df1=pd.DataFrame(df)
df1

#Finding missing values
df1.isnull().sum()

#replacing in gender
df1['gender'].mode()[0]
df1['gender'].value_counts()
#replacing missing value with mode
df1['gender'].fillna(df['gender'].mode()[0],inplace=True)
df1.isnull().sum()

#replacing missing values in test1 with mean 
df1['test1'].mean()
df1['test1'].value_counts()
df1['test1'].fillna(df1['test1'].mean(),inplace=True)
df1.isnull().sum()


#replacing missing values in test2 with mean
df1['test2'].mean()
df1['test2'].value_counts()
df1['test2'].fillna(df1['test2'].mean(),inplace=True)
df1.isnull().sum()
```

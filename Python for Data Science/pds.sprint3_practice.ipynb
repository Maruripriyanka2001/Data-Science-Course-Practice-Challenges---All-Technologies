```python
#TASK1:SORT THE DATA IN ASCENDING ORDER OF SALES AND DISPLAY THE RESULT
import pandas as pd
df=pd.read_csv(r"C:\Users\marur\Downloads\DS1_C4_S3_SalesData_Practice.csv")

#sorting 
df1=df.sort_values(by='Sales',ascending=True)
df1
```

```python
#TASK2: DISPLA ALL ROWS WHERE SALES AMOUNT IS >=45000
df2 = df[df['Sales'] >= 45000]
df2
```

```python
#TASK3 Filter the data separately for the states of California, Texas, and Montana and save them in three different data frames. 
#Sort the three data frames in increasing order of sale amount and display them separately.

#CALIFORNIA
df3=df[(df.State=='California')].sort_values(by='Sales',ascending=True)
df3
#TEXAS
df4=df[(df.State=='Texas')].sort_values(by='Sales',ascending=True)
df4
#MONTANA
df5=df[(df.State=='Montana')].sort_values(by='Sales',ascending=True)
df5
```

```python
#TASK4 :DISPLAY THE TOTAL SALES OF CALIFORNIA AND TEXAS AND MONTANA
print("total sales of california",df3['Sales'].sum())
print("total sales of Texas",df4['Sales'].sum())
print("total sales of Montana",df5['Sales'].sum())
```

```python
#TASK5: DISPLAY MOTHS WITH HIGHEST AND LOWEEST SALES AMOUNT IN THE STATE OF CALIFORNIA
# Filter the data for California
df6 = df[df['State'] == 'California']

# Find the maximum and minimum sales amounts
max_sales = df6['Sales'].max()
min_sales = df6['Sales'].min()

# Find the corresponding months for the maximum and minimum sales
max_sales_month = df6.loc[df6['Sales'] == max_sales, 'Month'].values[0]
min_sales_month = df6.loc[df6['Sales'] == min_sales, 'Month'].values[0]

# Display the results
print(f"The month with the highest sales amount is {max_sales_month} with sales of {max_sales}.")
print(f"The month with the lowest sales amount is {min_sales_month} with sales of {min_sales}.")






```

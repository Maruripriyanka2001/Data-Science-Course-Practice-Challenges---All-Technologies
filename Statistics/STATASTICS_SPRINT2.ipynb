```python
plt.boxplot([df['fare_amount']],labels=['fare_amount'])
plt.show()
```

```python
Q1 = df['fare_amount'].quantile(0.25)
Q3 = df['fare_amount'].quantile(0.75)
IQR = Q3 - Q1
lower_frence = Q1 - 1.5 * IQR
upper_frence = Q3 + 1.5 * IQR
df = df[(df['fare_amount'] >= lower_frence) & (df['fare_amount'] <= upper_frence)]
print(df.fare_amount)
```

```python
plt.boxplot([df['fare_amount']],labels=['fare_amount'])
plt.show()
```

```python
#TASK3 Draw different distribution plots for the following sample sizes:
#10, 15, 30, 200, 1000, 5000, 10000, 50000, 100000.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
fare_data = df['fare_amount']
sample_sizes = [10, 15, 30, 200, 1000, 5000, 10000, 50000, 100000]
plt.figure(figsize=(16, 12))
for i, size in enumerate(sample_sizes,start=1):
    sample = fare_data.sample(size, random_state=1, replace=True) 
    plt.subplot(3, 3, i)
    sns.histplot(sample, kde=True, bins=30, color='blue', edgecolor='black')
    plt.title(f"Sample Size: {size}")
    plt.xlabel("Fare Amount")
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

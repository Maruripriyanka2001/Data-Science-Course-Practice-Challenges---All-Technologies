```python

# Identified outliers for all the columns:
def outliers(df):
    for col in df.select_dtypes(include=['number']).columns:
        q1=df[col].quantile(0.25)
        q3=df[col].quantile(0.75)
        IQR=q3-q1
        LF=q1-1.5*IQR
        UF=q3+1.5*IQR
        outlier=df[(df[col]<LF) | (df[col]>UF)]
    return outlier

# call
alloutlier=outliers(df)
alloutlier
```

```python
# Removed outliers for all the columns:
def remove_outliers(df):
    for col in df.select_dtypes(include=['number']).columns:
        q1=df[col].quantile(0.25)
        q3=df[col].quantile(0.75)
        IQR=q3-q1
        LF=q1-1.5*IQR
        UF=q3+1.5*IQR
        outlier_removal=df[(df[col]>=LF) & (df[col]<=UF)]
    return outlier_removal
alloutlier=remove_outliers(df)
df 
```

```python
X=data.loc[:,['Age','AnyTransplants','AnyChronicDiseases']]
Y=data[['PremiumPrice']]
X
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.30,random_state=41)
LM=LinearRegression()
LM.fit(X_train,Y_train)
```

```python
b1=LM.coef_[0,0]
b2=LM.coef_[0,1]
b3=LM.coef_[0,2]
a=LM.intercept_
y=a+b1*.20+b2*.25+b3*20
y
```

```python
#Task 3 (cont'd)

# Build a linear regression model using the selected features.
#Extract and display the intercept and slope coefficients
#Evaluate the model performance using MSE, RMSE, MAE, and R-squared values.
```
